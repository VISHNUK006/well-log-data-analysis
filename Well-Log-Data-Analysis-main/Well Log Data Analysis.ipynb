1. Depth
Meaning: The vertical measurement from a reference point, typically the surface, to a specific point in the well. It is the key axis along which all other measurements are plotted.
2. RxoRt (Resistivity of the Flushed Zone)
Meaning: The resistivity of the zone closest to the borehole where the drilling fluid has flushed out most of the original formation fluids. It provides information about the formation's saturation and fluid content.
3. RLL3 (Deep Laterolog Resistivity)
Meaning: The resistivity measurement taken from the deeper part of the formation, farther from the borehole wall. This helps to understand the properties of the formation that are less affected by the drilling process.
4. SP (Spontaneous Potential)
Meaning: A measurement of the natural electric potential between the borehole and the surrounding formation. It is used to identify permeable zones and to estimate the salinity of the formation water.
5. RILD (Resistivity of the Deep Zone)
Meaning: This is another measure of deep resistivity, often used to assess the resistivity of the formation away from the borehole, where the influence of drilling fluids is minimal. It's similar to the deep laterolog resistivity but might be measured with a different tool or method.
6. MN (Macroscopic Neutron Porosity)
Meaning: A measurement of the hydrogen content in the formation, which is usually associated with the porosity of the rock. It is derived from neutron logging and helps estimate the porosity of the formation.
7. MI (Microlog Invasion)
Meaning: A resistivity log that measures the resistivity of the formation near the borehole, specifically where the drilling fluid has invaded the formation. It helps determine the extent of fluid invasion and the characteristics of the flushed zone.
8. MCAL (Microcaliper)
Meaning: A measurement of the borehole diameter, indicating the size and shape of the borehole at a very fine scale. It is useful for detecting borehole enlargements or washouts.
9. DCAL (Dual Caliper)
Meaning: A caliper log that measures the diameter of the borehole at two orientations (usually perpendicular to each other). It helps identify borehole stability issues, washouts, and the shape of the borehole.
10. RHOB (Bulk Density)
Meaning: A measurement of the density of the formation, including the pore spaces filled with fluid. It is used to estimate the formation's porosity and to identify lithology.
11. RHOC (Corrected Bulk Density)
Meaning: The bulk density reading corrected for borehole conditions, such as mudcake or borehole irregularities, to provide a more accurate estimate of the formation's true density.
12. DPOR (Density Porosity)
Meaning: The porosity derived from the bulk density measurement. It assumes a certain matrix density and compares it with the bulk density to estimate the amount of pore space in the formation.
13. CNLS (Compensated Neutron Log)
Meaning: A neutron porosity log that has been corrected for borehole size, lithology, and other factors to provide a more accurate porosity reading. It is often used in combination with the density log to determine the formation's porosity.
14. GR (Gamma Ray)
Meaning: A log that measures the natural gamma radiation emitted by the formation. High gamma ray readings typically indicate shale or clay-rich formations, while low readings suggest cleaner, non-shale formations like sandstones or carbonates.
These terms collectively help in evaluating the properties of subsurface formations, such as lithology, porosity, fluid content, and the presence of hydrocarbons.

#**2. Displaying a Well Plot with Matplotlib**

Setting up the logplot
We can quickly make a log plot using subplot2grid from matplotlib. This allows us to space out multiple plots (tracks) in an easy to understand way.


Each track is setup as an axis (ax for short)

```python
fig = plt.subplots(figsize=(10,10))

#Set up the plot axes
ax1 = plt.subplot2grid((1,3), (0,0), rowspan=1, colspan = 1)
ax2 = plt.subplot2grid((1,3), (0,1), rowspan=1, colspan = 1)
ax3 = plt.subplot2grid((1,3), (0,2), rowspan=1, colspan = 1)

ax1.plot("GR", "Depth", data = well, color = "green") # Call the data from the well dataframe
ax1.set_title("Gamma (API)") # Assign a track title
ax1.set_xlim(-10, 410) # Change the limits for the curve being plotted
ax1.set_ylim( 4808,195) # Set the depth range
ax1.grid() # Display the grid

ax3.plot("RHOB", "Depth", data = well, color = "red")
ax3.set_title("Density (g/cm3)")
ax3.set_xlim(1.25, 2.8)
ax3.set_ylim(4808,195)
ax3.grid()

ax2.plot("SP", "Depth", data = well, color = "purple")
ax2.set_title("SP (mV)")
ax2.set_xlim(-205, 75)
ax2.set_ylim(4808,195)
ax2.grid()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load the dataset (replace 'tntlog.csv' with the path to your actual dataset)
well = pd.read_csv('tntlog.csv')

# Ensure the dataset has the necessary columns: DCAL, SP, GR, RILD, RLL3, DPOR, CNLS, Depth
required_columns = ['Depth', 'DCAL', 'SP', 'GR', 'RILD', 'RLL3', 'DPOR', 'CNLS']
if all(col in well.columns for col in required_columns):
    # Create the figure and subplots with 4 columns and shared y-axis (Depth)
    fig, ax = plt.subplots(nrows=1, ncols=4, sharey=True, figsize=(25, 10))

    # Set the Y-axis to be inverted (depth increases downwards)
    for i in range(4):
     ax[i].invert_yaxis()


    # 1. Plot Caliper Log
    ax[0].plot(well['DCAL'], well['Depth'], label='Caliper', color='blue')
    ax[0].set_title('Caliper Log')
    ax[0].axhspan(3500, 5000, color='red', alpha=0.3, label='Wash-Out')
    ax[0].legend(loc='upper right')
    ax[0].set_ylim(5000,0)
    ax[0].grid()

    # 2. Plot Lithology Logs (SP and GR)
    shale_baseline = 60

    # Plot SP on the main axis
    ax1 = ax[1]
    ax1.plot(well["SP"], well["Depth"], color="green", linestyle='-', label="SP")
    ax1.set_xlabel("SP (mV)", color="green")
    ax1.set_xlim(well["SP"].min(), well["SP"].max())  # Auto x-axis range
    ax1.xaxis.label.set_color("green")
    ax1.tick_params(axis='x', colors="green")
    ax1.spines["top"].set_edgecolor("green")

    # Create a twin axis for Gamma Ray
    ax2 = ax1.twiny()
    ax2.plot(well["GR"], well["Depth"], color="brown", linestyle='-', label="Gamma Ray")
    ax2.set_xlabel("Gamma Ray (gAPI)", color="brown")
    ax2.set_xlim(well["GR"].min(), well["GR"].max())  # Auto x-axis range
    ax2.xaxis.label.set_color("brown")
    ax2.tick_params(axis='x', colors="brown")
    ax2.spines["top"].set_position(("axes", 1.02))  # Position above SP
    ax2.spines["top"].set_edgecolor("brown")
    ax2.axvline(x=shale_baseline, color="brown")  # Shale baseline for GR
    ax[1].set_title('Lithology Logs')
    ax[1].axhspan(3500, 5000, color='gold', alpha=0.3)
    ax[1].legend(loc='upper right')
    ax[1].grid()

    # 3. Plot Resistivity Logs (RILD and RLL3)
    ax[2].plot(np.log(well['RILD']), well['Depth'], label='RILD', color='blue')
    ax[2].plot(np.log(well['RLL3']), well['Depth'], label='RLL3', color='red')
    ax[2].set_title('Resistivity Logs')
    ax[2].axhspan(3500, 5000, color='gold', alpha=0.3)
    ax[2].axhspan(1500, 2500, color='green', alpha=0.3)
    ax[2].legend(loc='upper right')
    ax[2].grid()

    # 4. Plot Porosity Logs (DPOR and CNLS)
    ax[3].plot(well['DPOR'], well['Depth'], label='DPOR', color='red')
    ax[3].plot(well['CNLS'], well['Depth'], label='CNLS', color='purple')
    ax[3].set_title('Porosity Logs')
    ax[3].axhspan(3500, 5000, color='black', alpha=0.1)
    ax[3].axhspan(1500, 2500, color='green', alpha=0.3)
    ax[3].axhspan(2500, 3500, color='orange', alpha=0.6)
    ax[3].legend(loc='upper right')
    ax[3].grid()

    # Set a common Y-axis label (Depth)
    fig.text(0.04, 0.5, 'Depth (ft or m)', va='center', rotation='vertical')

    # Add a main title for the composite log
    plt.suptitle('Composite Well Log', fontsize=20)

    # Adjust spacing between subplots
    plt.subplots_adjust(wspace=0.3)

    # Display the plot
    plt.show()
else:
    print(f"Dataset is missing one or more required columns: {required_columns}")
```
